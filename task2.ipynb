```python
df['sepal_length'].hist(bins=10)

plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
df['sepal_width'].hist(bins=7)

plt.title('Distribution of Sepal width')
plt.xlabel('Sepal width')
plt.ylabel('Frequency')
plt.show()
```

```python
df['petal_length'].hist(bins=7)

plt.title('Distribution of petal Length')
plt.xlabel('petal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
df.boxplot(column=['sepal_length','sepal_width','petal_length','petal_width'])

plt.title('Boxplot of Numerical Features')
plt.show()
```

```python
plt.scatter(
    df['sepal_length'],
    df['petal_length']
)

plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(6,4))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True
)

plt.title('Correlation Matrix')
plt.show()
```

Negative correlation values indicate an inverse relationship between variables. As one feature increases, the other tends to decrease. Positive correlation values indicate that both features tend to increase together.
